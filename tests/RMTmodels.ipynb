```julia
using Revise
#using Random, Distributions
using CSV
using DataFrames
using StatsBase
using CairoMakie

include("../src/main.jl")
import .SpectralStatistics as stats
```

```julia
poi = stats.Poisson()
goe = stats.GOE()
gue = stats.GUE()
gse = stats.GSE()

models = [poi, goe, gue, gse]
```

```julia
#level spacings of order n
n = 1
min_x = 0.0
max_x = 5.0
grid = 200
x = collect(LinRange(min_x, max_x, grid))

fig = Figure()
ax = fig[1, 1] = Axis(fig)

for m in models
    y = stats.level_spacing_pdf(m,x; n=n) 
    lin = lines!(ax, x, y)
end

fig
```
